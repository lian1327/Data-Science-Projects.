```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
countries = pd.read_csv('countries.csv')
```

```python
countries
```

```python
countries.info()
```

```python
countries.describe()
```

```python




# Filter the data for the year 1952 and for the continents Asia and Africa
filtered_data = countries[(countries['year'] == 1952) & (countries['continent'].isin(['Asia', 'Africa']))]

# Separate life expectancy data for Asia and Africa
asia_life_expectancy = filtered_data[filtered_data['continent'] == 'Asia']['population']
africa_life_expectancy = filtered_data[filtered_data['continent'] == 'Africa']['population']

# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns for comparison

# Plot histogram for Asia
axes[0].hist(asia_life_expectancy, bins=15, color='blue', edgecolor='black', alpha=0.7)
axes[0].set_title('Life Expectancy in Asia (1952)', fontsize=14)
axes[0].set_xlabel('Life Expectancy (Years)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].grid(True, linestyle='--', alpha=0.5)

# Plot histogram for Africa
axes[1].hist(africa_life_expectancy, bins=15, color='green', edgecolor='black', alpha=0.7)
axes[1].set_title('Life Expectancy in Africa (1952)', fontsize=14)
axes[1].set_xlabel('Life Expectancy (Years)', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].grid(True, linestyle='--', alpha=0.5)

# Add a main title for the entire figure
fig.suptitle('Comparison of Life Expectancy in Asia and Africa (1952)', fontsize=16)

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space for the main title

# Show the plot
plt.show()
```
