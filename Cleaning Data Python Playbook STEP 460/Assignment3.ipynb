<h3>Checking Data (cleaning) </h3>

```python
estate.isnull().sum()
```

```python
### 1. Normalize the data in the houseAge column.
# Normalization formula: (x - min(x)) / (max(x) - min(x))
estate['houseAgeStandardized'] = (estate['houseAge'] - estate['houseAge'].min()) / (estate['houseAge'].max() - estate['houseAge'].min())

# Comment: This creates a new column 'houseAgeStandardized' where values are scaled between 0 and 1.

estate
```

```python
### 2. Drop the numberOfConvenienceStores column and apply this change to the dataset.
estate.drop(columns=['numberOfConvenienceStores'], inplace=True)
estate
# Comment: The 'numberOfConvenienceStores' column is removed from the dataset permanently using `inplace=True`.
```

```python
### 3. Change the transaction column name to transactionDate.
estate.rename(columns={'transaction': 'transactionDate'}, inplace=True)

# Comment: Renames the 'transaction' column to 'transactionDate' for clarity.
estate
```

```python

### 4. Use .loc[] to display all rows from 0-10.
print(estate.loc[0:10])

# Comment: Displays rows indexed from 0 to 10 (inclusive) using `.loc[]`.

```

```python
### 5. Use .iloc[] to display the first 10 rows.
print(estate.iloc[0:10])

# Comment: Displays the first 10 rows (indexed from 0 to 9) using `.iloc[]`.

```

```python
### 6. Find and remove any duplicate data rows.
# Identify duplicates
duplicates = estate.duplicated()
print("Duplicate Rows:\n", estate[duplicates])

# Remove duplicates
estate.drop_duplicates(inplace=True)

# Comment: Identifies and removes duplicate rows from the dataset.

```

```python
### 7. Find any missing values and fill them with the mean.
# Check for missing values

estate.isnull().sum()

estate['housePriceOfUnitArea'] = estate['housePriceOfUnitArea'].fillna(estate['housePriceOfUnitArea'].mean())

estate.isnull().sum()
```
