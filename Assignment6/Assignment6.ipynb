```python

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Separate features (independent variables) and target (dependent variable)
# Here, 'charges' is the target variable, and the rest are features
X = df.drop('charges', axis=1)  # Features
y = df['charges']  # Target

# Apply LabelEncoder to categorical data ('smoker')
# LabelEncoder converts categorical text data into numerical values
label_encoder = LabelEncoder()
X['smoker'] = label_encoder.fit_transform(X['smoker'])  # Encode 'smoker' column (yes/no -> 1/0)

# Apply OneHotEncoder to categorical data ('children')
# OneHotEncoder converts categorical integer data into a one-hot numeric array
# We use ColumnTransformer to apply OneHotEncoder only to the 'children' column
column_transformer = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(), ['children'])  # Apply OneHotEncoder to 'children'
    ],
    remainder='passthrough'  # Keep other columns unchanged
)
X = column_transformer.fit_transform(X)

# Split the data into training and testing sets
# Set the test size to 15% (0.15) as per the requirement
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model using Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Print the coefficients of the linear regression model
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)

# Optional: Display actual vs predicted values for comparison
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(comparison.head())
```
